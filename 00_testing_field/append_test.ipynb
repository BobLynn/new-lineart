```python
a = [1,2,3,4] 
b = a[:]         # b 為 a 的分身

print(f"b: {b}") # 先輸出b看看
b.append(5)      # 將5加入b

# 比較a跟b
print(f"a: {a}") # 比較a跟b
print(f"b: {b}")

print("------------------")

a = [1,2,3,4] 
b = a.copy()     # b 為 a 的分身

print(f"b: {b}") # 先輸出b看看
b.append(5)      # 將5加入b

# 比較a跟b
print(f"a: {a}") # 比較a跟b
print(f"b: {b}")

print("------------------")

a = [1,2,3,4] 
b = list(a)      # b 為 a 的分身

print(f"b: {b}") # 先輸出b看看
b.append(5)      # 將5加入b

# 比較a跟b
print(f"a: {a}") # 比較a跟b
print(f"b: {b}")

print("------------------")

a = [1,2,3,4] 
b = [i for i in a] # b 為 a 的分身

print(f"b: {b}") # 先輸出b看看
b.append(5)      # 將5加入b

# 比較a跟b
print(f"a: {a}") # 比較a跟b
print(f"b: {b}")
```
